```python
import requests
from simplified_scrapy.simplified_doc import SimplifiedDoc
import pandas as pd
```

```python
data_info = []
```

```python
def data_select(url):
    
  source = requests.get(url).text
  doc = SimplifiedDoc(source)
  infos = doc.getElements('div', attr='class', value='favorite dui-menu-item')

  for info in infos:
    if info:
        p_url = info['data-item-url']
        product = info['data-item-name']
        price = info['data-price']
        score  = info['data-review-avg']
        data_info.append([p_url, product, price, score])
```

```python
data_select('https://search.rakuten.co.jp/search/mall/air+purifier/')
data_select('https://search.rakuten.co.jp/search/mall/air+purifier/?p=2')
```

```python
df = pd.DataFrame(data_info)
df.columns = ['url', 'description', 'price', 'score']
df.head(20)
```

```python
#####################################################################
from win32com.client import Dispatch
import os, fnmatch, sys

#####################################################################
def count_and_display_file_name(file_list):
    print(f"총 파일수 = {len(file_list)}")

    for i, file in enumerate(file_list):
        print(f"[{i}] : {file}")
#####################################################################
# D:\Kiwoom\pyQt5
old_path  = 'C:\\Kiwoom\\pyQt5\\공익법인_결산서류\\'
new_path  = 'C:\\Kiwoom\\pyQt5\\수정_공익법인_결산서류\\'

#####################################################################
file_list = fnmatch.filter(os.listdir(old_path), '*.xls')
count_and_display_file_name(file_list)
#sys.exit()

#####################################################################
#################   Start main program from here ####################
#####################################################################
for i in range(len(file_list)):

    source_path = old_path + file_list[i]
    output_path = new_path + file_list[i] + "x"
    
    print(source_path)
    excel = Dispatch("Excel.Application")
    wb = excel.Workbooks.Open(Filename=source_path) 
    ws = wb.Worksheets(1)
    
    print("="*80)
    law_edit = ws.Range("K1").value
    
    if law_edit == None:
        print(ws.Range("A1").value)
        company  = ws.Range("C6").value
        ceo      = ws.Range("C7").value
        birthday = ws.Range("R7").value
        address  = ws.Range("C8").value
        phone_no = ws.Range("R8").value
        homepage = ws.Range("C9").value
        email    = ws.Range("R9").value   
    else:      
        print(ws.Range("A1").value, ws.Range("K1").value)
        company  = ws.Range("E6").value
        ceo      = ws.Range("E7").value
        birthday = ws.Range("X7").value
        x = dir(ws.Range("X7").value)
        y = type(birthday)
        address  = ws.Range("E8").value
        phone_no = ws.Range("X8").value
        homepage = ws.Range("E9").value
        email    = ws.Range("X9").value
    print(f"법인회사 : {company}")
    print(f"대표자   : {ceo}")
    print(f"설립일자 : {birthday}")
    print(f"소재지   : {address}")
    print(f"전화번호 : {phone_no}")
    print(f"홈페이지 : {homepage}")
    print(f"전자주소 : {email}")
    print("-"*80)
    
    wb.SaveAs(output_path, FileFormat = 51)    #FileFormat = 51 is for .xlsx extension 

    wb.Close()                                 #FileFormat = 56 is for .xls extension 
    print(output_path)

    excel.Application.Quit()
    
#####################################################################
##################    End main program here   #######################
#####################################################################
file_list = fnmatch.filter(os.listdir(new_path), '*.xlsx')
count_and_display_file_name(file_list)
#####################################################################
```
